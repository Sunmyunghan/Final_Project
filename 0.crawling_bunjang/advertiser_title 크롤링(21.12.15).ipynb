# 실습

```python
# 빈 딕셔너리 생성
dic_ = {
    'user_id': [],
    'title': []
}
```

```python
# 크롬창 열기
driver = webdriver.Chrome('./chromedriver.exe')

title = []

for i in user_id:
    # url정의
    url = f'https://m.bunjang.co.kr/shop/{i}/products'
    driver.get(url)
    # xpath를 이용한 데이터 크롤링
    try:
        title = driver.find_element_by_xpath('//*[@id="root"]/div/div/div[4]/div/div[1]/div/div[2]/div[1]/div[1]').text
    except:
        title = "none"
        pass
    
    dic_["user_id"].append(i)
    dic_["title"].append(title)
```

```python
len(dic_["user_id"]), len(dic_["title"])
```

```python
# DataFrame화
df = pd.DataFrame(dic_)
df
```

```python
# "title"이 존재하지 않는 data
df[df["title"]=="none"]
```

```python
# data값 변경
df.loc[202, "title"] = 'Leopard7'
```

```python
df_title = pd.merge(advertiser, df, how='left', on="user_id")
```
